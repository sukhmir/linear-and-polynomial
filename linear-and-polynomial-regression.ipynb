#### 3.1 Predicting Student Marks Objective: The objective of this lab task is to build a linear and polynomial regression model to predict student marks based on the number of courses taken and the number of study hours. Dataset: The dataset consists of records of students’ marks along with the number of courses theyhave taken and the number of hours they have studied. The dataset is available in the followingURL


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\Shoaib sukhmir\\AI_Data_science\\Student_Marks.csv")
df
```

```python
x=pd.array(df["number_courses"])
y=pd.array(df["time_study"])
#plt.plot(number_courses,time_study,color="red")
```

```python
x
```

```python
x = x.reshape(-1, 1)
from sklearn.linear_model import LinearRegression
linear=LinearRegression()
linear.fit(x,y)
```

```python
#predict marks
predict_marks=linear.predict(x)
print("Time",x)
print("Actual marks",y)
print("predicted marks",predict_marks)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color="red",label="Actual Data points")
plt.plot(x,predict_marks,color="blue",label="Predicted plot")
```

```python
from sklearn.metrics import  mean_squared_error
mse=mean_squared_error(y,predict_marks)
rmse=np.sqrt(mse)
print("Mean square error",mse)
print("Root mean square error",rmse)
```

```python
#polymomial regression
from sklearn.preprocessing import PolynomialFeatures
hours_poly=PolynomialFeatures(degree=7).fit_transform(x)
poly_model=LinearRegression()
poly_model.fit(hours_poly,y)
```

```python
predict_marks=poly_model.predict(hours_poly)
plt.scatter(x,y,color="red",label="Actual Data points")
plt.plot(x,predict_marks,color="blue",label="predicted graph")
```

```python
print("Hours", x)
print("Actual Marks", y)
print("Predicted Marks", predict_marks)
```
